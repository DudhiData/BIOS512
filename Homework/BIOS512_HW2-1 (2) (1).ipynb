# Question 9 Answer
The counter2 function works while the counter1 function does not work. 
The difference in how R processes the two functions arrises when counter2 uses <<-, which finds the nearest enclosing binding for val and adds the step value to its own value. When counter1 doesn't use <<-, the value of val in the tertiary environment changes but that isn't reflected in the secondary environment, so the counter doesn't increase the start value at all. 
    

```r

counter1 <- function(start, step){
    val <- start;
    function(){
        old_val <- val;
        val <- val + step;
        old_val;
    }
}
counter_from_1 <- counter1(1,1);
counter_from_1()
counter_from_1()

counter2 <- function(start, step){
    val <- start;
    function(){
        old_val <- val;
        val <<- val + step;
        old_val;
    }
}
counter_from_1 <- counter2(1,1);
counter_from_1()
counter_from_1()
```
