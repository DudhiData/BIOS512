b)              
The variables have vastly different variances so scaling would allow us to balance variable influence. Centering the data means shifting the origin of the PCA to the mean, thus the mean will not create a bias in the PCA.

```r
wine_pca <- prcomp(wine_sc)
summary(wine_pca)
```

b)               
PC1 explains 38.21% of the variance, while PC2 explains 18.36% of the variance. To get this info, we use the prcomp() function on our scaled and centered dataset and then call summary() on the result of that. 

c)        
We do the PCA first to find the set of principal components that we will use to drop the rest of the features and thus reduce the dimensionality of the dataset.

```r
# d) 
wine_pca
```

```r
wine_pc <- as.data.frame(wine_pca$x) %>% select(PC1, PC2) %>% bind_cols(class = wine_sc$class)
ggplot(wine_pc, aes(PC1, PC2)) + geom_point(aes(color = factor(class))) + labs(color = "Wine Class")
```

f)     
We see 3 distinct clusters. In the context of wine classes, this tells us that the principal components are significant enough to effectively differentiate distinctions between the 3 classes. 

g)       
A vector space is a set V along with a field F that satisfies the necessary vector space axioms at all times. However, a manifold is merely a set that has the property of resembling a vector space only locally. In terms of TSNE, our dataset is our vector space, and the manifold is the underlying lower-dimensional structure on which this dataset can be defined on, so what TSNE does is find a way to map from the global vector space to the underlying local one (the manifold).

## Question 4
#### a) Perform T-SNE
Set `seed = 123`.  
*Hint:* Subset your PCA results to PC1–PC10, add the class variable back in, remove duplicates, then perform T-SNE.

#### b) Plot the results in 2D
*Hint:* Convert your T-SNE results to a tibble and add back the class variable from your scaled data set using a `mutate()` statement. Then, you can use `color = factor(class)` in your `ggplot` statement.

#### c) Why didn't we stop at PCA?


#### d) What other types of data does this workflow make sense for?



```r
# A
set.seed(123)
wine_tsne <-as.data.frame(wine_pca$x) %>% select(c(PC1:PC10)) %>% bind_cols(class = wine_pc$class) %>% distinct()
results <- Rtsne(wine_tsne, dims = 2, perplexity = 30, verbose = FALSE, check_duplicates = FALSE)
# B
wine_data <- as_tibble(results$Y) %>% mutate(class = wine_sc$class)
ggplot(wine_data, aes(V1, V2)) + geom_point(aes(color = factor(class))) + labs(color = "Wine Class")
```

c)           
PCA only works for linear relationships, while TSNE is great for identifying non-linear relationships. So, it would make sense to keep going to possibly expose some hidden relationships that PCA wouldn't have shown us.

d)          
Genetic data, image data, linguistic data, and many other types of data that have high dimensionality with a lower dimensional local structure would make sense with this sort of workflow.
