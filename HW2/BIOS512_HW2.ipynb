## Nonstandard Evaluation


### Question 1
Imagine we have a data frame called `data`, with a `type` column. Which one works and why?\
Function 1:
```
group_and_tally <- function(df, column){
    df %>% group_by({{ column }}) %>% tally();
}
group_and_tally(data, type);
```

Function 2:
```
group_and_tally <- function(df, column){
    df %>% group_by(column) %>% tally();
}
group_and_tally(data, type);
```

Function 1 works while Function 2 doesn't because in the environment of the function, when you call the group_by function on 'df' and pass in 'column', the issue becomes that R tries to look for the column named 'column' in df, which will cause an error in Function 2. In function 1, the {{}} makes use of nonstandard evaluation and passes the column parameter instead of the column name 'column' into the group_by function, which will correctly perform group_by on 'df'.
